```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Hackathon DS/archive(1)/DataSet_Misinfo_TRUE.csv')
```

```python
df.dropna(inplace=True)
df
```

```python
import re # Remove punctuation

df['text_processed'] = df['text'].map(lambda x: re.sub('[,\.!?]', '', x)) 
df['text_processed'] = df['text_processed'].map(lambda x: x.lower()) 
df['text_processed'][0]
```

```python
df[:100]
```

```python
len(df.text_processed.values.tolist()[:100])
```

```python
sw = ['from', 'subject', 're', 'edu', 'use', 'us', 'said', 'mr', 'even', 'also', 'get', 'would',
      'like', 'first', 'know', 'make', 'could', 'two', 'may', 'many', 'ms', 'one']
```

```python
import gensim
from gensim.utils import simple_preprocess
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words.extend(sw)

def sent_to_words(sentences):
  for sentence in sentences:
    # deacc=True removes punctuations
    yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))
def remove_stopwords(texts):
  return [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in texts]

data = df.text_processed.values.tolist()#[:100]
data_words = list(sent_to_words(data))# remove stop words
data_words = remove_stopwords(data_words)
print(data_words[:1][0][:30])

# data_words = list(sent_to_words(data))# remove stop words
# data_words = remove_stopwords(data_words)
# print(data_words[:1][0][:30])
```

```python
import gensim.corpora as corpora # Create Dictionary

id2word = corpora.Dictionary(data_words) # Create Corpus
texts = data_words # Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts] # View
print(corpus[:1][0][:30])
```

```python
from pprint import pprint # number of topics
num_topics = 15 # 10 Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics) # Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```
